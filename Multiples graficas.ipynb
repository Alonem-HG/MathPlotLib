# Renderizar 

* Que y cuando renderizar 


```python
def plotPendiente(X,Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:],Ys/Xs)
```

```python
X = np.linspace(-3,3,100)
Y = np.exp(-X ** 2)
plt.plot(X,Y)
plotPendiente(X,Y)
plt.show()


```

# Plotting curves from file data

* Plot datos de un archivo .txt 
* Diferentes formas de cargar archivos
* Diferente formas de graficar datos de archivos 

```python
X,Y = [],[]
for line in open('data.txt','r'):
    valores = [float(s) for s in line.split(",")]
    X.append(valores[0])
    Y.append(valores[1])
plt.plot(X,Y)
plt.show()
```

```python
with  open('data.txt','r') as f:
    X,Y,Z,A,B,C = zip(*[[float(s) for s in line.split(",")] for line in f ])
plt.plot(X,A)

plt.show()
```

```python
data = np.loadtxt("data.txt", delimiter=',')
plt.plot(data[:,0],data[:,1])
plt.show()
```

```python
data = np.loadtxt("data.txt", delimiter=',')
for columna in data.T:
    plt.plot(data[:,0],columna)
plt.show()
```
