# Personalizacion de Colores

* rc objeto
* propiedades de ese objeto 

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
from cycler import cycler
mpl.rc('lines', linewidth = 2.)
mpl.rc('axes', facecolor = 'k', edgecolor = 'r')
mpl.rc('axes', prop_cycle=(cycler('color', ['b', 'g', 'y'])))
mpl.rc('xtick', color = 'w')
mpl.rc('ytick', color = 'w')
mpl.rc('text', color = 'r')
mpl.rc('font', size = 12)
mpl.rc('grid', color = 'm')
mpl.rc('grid', linewidth   = 1.0)

mpl.rc('figure', facecolor = 'r', edgecolor ='k')
# https://matplotlib.org/1.4.0/users/customizing.html

X = np.linspace(0, 7, 1024)
plt.title(" grafica ")
plt.plot(X, np.sin(X))
plt.plot(X, np.cos(X))
plt.plot(X, np.cos(X)+np.sin(X))
plt.grid()
plt.show()
```

```python
X = np.linspace(0, 7, 1024)
plt.plot(X, X**2)
plt.plot(X, X**3)
plt.plot(X, X**4)
```
