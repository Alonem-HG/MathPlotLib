# Personalizando grafica de barras

Debido al gran uso de de barras y presentaciones, una buena presentacion de colores simpre debe estar presente.


```python
import numpy as np
import matplotlib.pyplot as plt
pob_mujeres = np.array([5., 30., 45., 22.]) 
pob_hombres = np.array( [5., 25., 50., 20.])
X = np.arange(4)
plt.barh(X, pob_mujeres, color = '#46C924', edgecolor='0.2')
plt.barh(X, -pob_hombres,color = 'r', edgecolor='0.1')
plt.show()
```

*  pyplot.bar()
* pyplot.barh()
* pyplot.scatter()   similares en color y edgecolor

# mas con barras

```python
valores = np.random.randint(99, size = 50)
valores.sort()
set_de_colores = ('r', 'g', '.50', '#C4B693')
lista_de_colores = \
[set_de_colores[(len(set_de_colores) * val) // 100] for val in valores]
plt.bar(np.arange(len(valores)), valores, color = lista_de_colores)
plt.show()
```

# Colores para diagramas de Caja y bigotes

```python
valores = np.random.randn(100)
b = plt.boxplot(valores)
for nombres, lista_de_lineas in b.items():
    print(nombres)
    for linea in lista_de_lineas:
        linea.set_color('0.1')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
valores = np.random.randn(100)
b = plt.boxplot(valores)
for nombres, lista_de_lineas in b.items(): #regresa el diccionario de items
    print(nombres)
    #print(lista_de_lineas)
    for linea in lista_de_lineas:
        if nombres =='whiskers' :linea.set_color('g') 
        if nombres =='caps' :linea.set_color('r') 
        if nombres =='boxes' :linea.set_color('b') 
        
plt.show()
```

# colormaps para graficos scatter 

* mapas de colores
* matplotib.cm
* este modulo provee la funcion para crear mapas de colores



```python
import matplotlib.cm as cm
```

```python
N = 256
angulo = np.linspace(0, 8 * 2 * np.pi, N)
radio = np.linspace(.5, 1., N)
X = radio * np.cos(angulo) # representacion polar de un circulo
Y = radio * np.sin(angulo)
plt.scatter(X, Y, c = angulo, cmap = cm.PuOr,edgecolor='g') #edgecolor='0.2'
plt.show()
```

# Controlando el patron de lineas y el grosor de linea

si tenemos documentos en solo blanco y negro necesitaremos los niveles de grises y usando varios patrones de linea nos da una variedad

```python
def pdf(X, mu, sigma):
    a = 1. / (sigma * np.sqrt(2. * np.pi))
    b = -1. / (2. * sigma ** 2)
    return a * np.exp(b * (X - mu) ** 2)
X = np.linspace(-6, 6, 1024)
plt.plot(X, pdf(X, 0., 1.), color = '0.5', linestyle = 'solid')
plt.plot(X, pdf(X, 0., .5), color = '0.2', linestyle = 'dotted')
plt.plot(X, pdf(X, 0., .25), color = 'k', linestyle = 'dashdot')
plt.show()
```
