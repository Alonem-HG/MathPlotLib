## Definiendo nuestros propios colores

```python
# grafica de densidad
def pdf(X, mu, sigma):
    a = 1. / (sigma * np.sqrt(2. * np.pi))
    b = -1. / (2. * sigma ** 2)
    return a * np.exp(b * (X - mu) ** 2)
X = np.linspace(-6, 6, 1000)
for i in range(5):
    samples = np.random.standard_normal(50)
    mu, sigma = np.mean(samples), np.std(samples)
    plt.plot(X, pdf(X, mu, sigma), color = '0.7')
plt.plot(X, pdf(X, 0., 1.), color = 'k') #negro
plt.show()
```

# Color en scatter  
* cambio de colores en scatter


```python
A = np.random.standard_normal((100, 2))
A += np.array((-1, -1)) # CENTRO de la distribucion en <-1, -1>
B = np.random.standard_normal((100, 2))
B += np.array((1, 1)) # CENTRO de la distribucion en at <1, 1>
plt.scatter(A[:,0], A[:,1], color = '.25')
plt.scatter(B[:,0], B[:,1], color = 'y')
plt.show()
```

# Graficando desde base de datos IRIS 
* graficos desde base datos ( tutor leer archivos)
* scatter , etiquestas , nombres y colores


```python
? np.loadtxt
```

```python
set_de_etiquetas = (
b'Iris-setosa',
b'Iris-versicolor',
b'Iris-virginica',
)
def leer_label(label):
    return set_de_etiquetas.index(label)
data = np.loadtxt('iris.data',
        delimiter = ',',
        converters = { 4 : leer_label })
set_de_colores = ('.00', 'r', 'g') # ahora podemos controlar como usar los colores de manera personalizada
lista_de_colores = [set_de_colores[int(label)] for label in data[:,4]]
plt.scatter(data[:,0], data[:,1], color = lista_de_colores)
plt.show()
```
