```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = np.array([[5., 30., 45., 22.],
                [5., 25., 50., 20.],
                [1., 2., 1., 1.]])

lista_de_colores = ['b','r','g']
X= np.arange(data.shape[1])
for i in range (data.shape[0]):
    plt.bar(X, data[i],
           bottom = np.sum(data[:i], axis = 0),
           color = lista_de_colores[i % len(lista_de_colores)])
plt.show()
```

# Barras de espalda con espalda
* tallo y hojas

```python
pob_mujeres = np.array([5., 30., 45., 22.]) 
pob_hombres = np.array( [5., 25., 50., 20.])
X = np.arange(4)
plt.barh(X, pob_mujeres, color = 'g')
plt.barh(X, -pob_hombres,color = 'r')
plt.show()
```

# Grafica de triangulaciones
* mejor ayuda visual de la localizacion de los puntos o datos

```python
data = np.random.rand(100, 2)
```

```python
import matplotlib.tri as tri
```

```python
triangulos  = tri.Triangulation(data[:,0],data[:,1])
plt.triplot(triangulos)
plt.show()
```
