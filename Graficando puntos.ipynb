# Graficando multiples barras 

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]
X= np.arange(4)
plt.bar(X + 00.00,data[0],color = 'b',width = 0.20)
plt.bar(X + 0.20 ,data[1],color = 'r',width = 0.20)
plt.bar(X + 0.40 ,data[2],color = 'g',width = 0.20)
```

* vamos a ser mas intuitivos

```python
data = [[5., 25., 50., 20.,4],
        [4., 23., 51., 17.,3],
        [6., 22., 52., 19.,6],
        [6., 22., 52., 19.,6]]
lista_de_colores = ['r','b','g']
brecha = 0.8/len(data)
for i,fila in enumerate(data):
    X = np.arange(len(fila))
    plt.bar(X + i*brecha , fila , width = brecha , 
           color = lista_de_colores[i % len(lista_de_colores)])
```

# Graficando barras verticales apiladas
* stacked bar charts

```python
A = [5., 30., 45., 22.]
B = [5., 25., 50., 20.]
X = range(4)
plt.bar(X,A,color='b')
plt.bar(X,B,color='g',bottom=A)
plt.show()

```
