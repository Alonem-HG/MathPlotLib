```python
import numpy as np
from matplotlib import pyplot as plt
T = np.linspace(-np.pi, np.pi, 1024) # entre pi y -pi
tama_cuadric = (4, 2)
```

```python
plt.subplot2grid(tama_cuadric, (0, 0),rowspan= 3, colspan = 1) # espacio de fila y columna
plt.plot(np.sin(2 * T), np.cos(0.5 * T), c = 'k') 
plt.subplot2grid(tama_cuadric, (0, 1), rowspan = 3, colspan = 1)
plt.plot(np.cos(5 * T), np.sin(T), c = 'r')
plt.subplot2grid(tama_cuadric, (3, 0), rowspan=1, colspan=3)
plt.plot(np.cos(5 * T), np.sin(7 * T), c= 'k')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
T = np.linspace(-np.pi, np.pi, 256)
fig,(ax0, ax1)= plt.subplots(ncols =2)
ax0.plot(np.sin(2 * T), np.cos(0.5 * T), c = 'k' )
ax1.plot(np.cos(3 * T), np.sin(T), c = 'k')
plt.show()
```

# Escalar ambos ejes igualmente

```python
T = np.linspace(0, 2 * np.pi, 1024)
plt.plot(2. * np.cos(T), np.sin(T), c = 'k', lw = 3.)
plt.axes().set_aspect(2)
plt.show()
```

```python
#rango de los ejes

X = np.linspace(-6, 6, 1024)
plt.ylim(-.5, 1.5) #rango del eje
plt.plot(X, np.sinc(X), c = 'k')
plt.show()
```

```python
# aspect ratio

X = np.linspace(-6, 6, 1024)
Y1, Y2 = np.sinc(X), np.cos(X)
plt.figure(figsize=(10.24, 2.56)) #4:1
plt.plot(X, Y1, c='k', lw = 3.)
plt.plot(X, Y2, c='.75', lw = 3.)

plt.show()
```

```python
# insertando Subfiguras
X = np.linspace(-6, 6, 1024)
Y = np.sinc(X)
X_detail = np.linspace(-2, 2, 1024)
Y_detail = np.sinc(X_detail)
plt.plot(X, Y, c = 'k')
sub_axes = plt.axes([.6, .6, .25, .25])
sub_axes.plot(X_detail, Y_detail, c = 'k')
#plt.setp(sub_axes)
plt.show()
```

Usar escala logaritmica

```python
X = np.linspace(1, 10, 1024)
plt.figure(dpi=120)
plt.yscale('log')
plt.plot(X, X, c = 'k', lw = 2., label = r'$f(x)=x$')
plt.plot(X, 10 ** X, c = '.75', ls = '--', lw = 2., label =r'$f(x)=e^x$')
plt.plot(X, np.log(X), c = '.95', lw = 2., label = r'$f(x)=\log(x)$')
plt.legend()
plt.show()
#plt.savefig('filename.png', dpi=300)
```
